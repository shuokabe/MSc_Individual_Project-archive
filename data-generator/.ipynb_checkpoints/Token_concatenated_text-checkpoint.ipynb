# Read token data

```python
import pandas as pd
```

```python
train_token = pd.read_csv('QE-2018-task4-titles-categories-and-ids_training.txt', sep='\t', header=None)
train_token.columns = ["doc_no", "asin", "title", "token"]
train_token.head()
```

```python
# Check how many products have no tokens
train_token.isna().sum()
```

```python
# Replace NaNs with empty strings
train_token = train_token.fillna('')
print(train_token.shape)
train_token.head(10)
```

```python
# Split the token information with respect to '@@''
train_token["token_list"] = train_token.token.str.split('@@') 
train_token.head(10)
```

```python
# Count the number of sub-categories per product
train_token["len_token_list"] = train_token.token_list.str.len()
train_token.head(10)
```

```python
# See the distribution of the number of sub-categories
train_token.len_token_list.value_counts()
```

```python
# Standardise the number of tokens to 6 (maximum number of sub-categories)
def pad_list(data, max_len=6):
    data['token_list_pad'] = ''
    data['string_token_list_pad'] = ''
    for index, row in data.iterrows():
        pad_number = max_len - row.len_token_list
        data.at[index, 'token_list_pad'] = row.token_list + ([''] * pad_number)
        data.at[index, 'string_token_list_pad'] = ' @ '.join(data.at[index, 'token_list_pad'])
    return data

train_token = pad_list(train_token, 6)
train_token.head(10)
```

```python
# Load the other datasets
# Dev dataset
dev_token = pd.read_csv('QE-2018-task4-titles-categories-and-ids_dev.txt', sep='\t', header=None)
dev_token.columns = ["doc_no", "asin", "title", "token"]
dev_token = dev_token.fillna('')

dev_token["token_list"] = dev_token.token.str.split('@@') 
dev_token["len_token_list"] = dev_token.token_list.str.len()
dev_token = pad_list(dev_token, 6)


# Test dataset
test_token = pd.read_csv('QE-2018-task4-titles-categories-and-ids_test.txt', sep='\t', header=None)
test_token.columns = ["doc_no", "asin", "title", "token"]
test_token = test_token.fillna('')

test_token["token_list"] = test_token.token.str.split('@@') 
test_token["len_token_list"] = test_token.token_list.str.len()
test_token = pad_list(test_token, 6)
```

```python
# Read the file containing the document ID for each sentence in the sentence-level QE data
train_doc_id = pd.read_csv('train.doc_id', header=None)
train_doc_id.columns = ['doc_id']
train_doc_id.head(10)
```

```python
# The train source
with open('../train.src_sent', 'w') as outfile:
    i = 0
    with open('train.src', 'r') as infile:
        for line in infile: 
            outfile.write(train_sent_merge.token[i] + "@@" + line)
            i += 1
      
```
