```python
# Read the visual features (code from the provider of the data: http://jmcauley.ucsd.edu/data/amazon/)

import struct

def readImageFeatures(path):
    f = open(path, 'rb')
    while True:
        asin = f.read(10)
        print(asin)
        if asin == '': 
            break
        feature = []
        for i in range(4096):
            feature.append(struct.unpack('f', f.read(4)))
        yield asin, feature
        #return asin, feature
```

```python
# The visual features are stored in a generator object
image_feature = readImageFeatures("image_features_Sports_and_Outdoors.b")   
print(image_feature)
```

```python
import numpy as np

# Add the visual feature to the data
def findVisualFeature(data, feature_gen):
    desired_prod = list(data.asin)
    count = 0
    for gen_element in feature_gen:
        vis_asin, vis_feature = gen_element[0].decode("utf-8"),  gen_element[1] # From generator
        #print("Test of type", vis_asin == "B0007RTC3C")
        data_index = data.index[data.asin == vis_asin].to_list() # List of indices (normally one element) where doc_asin matches gen_asin
        if len(data_index) > 1: # Duplicate indices
            print("Duplicate:", data_index)
            data.iloc[data_index[0], data.columns.get_loc("feature")] = gen_element[1]
            data.iloc[data_index[1], data.columns.get_loc("feature")] = gen_element[1]
            #continue
            count += 2
            print(count)
        elif len(data_index) == 1:
            #print(data_index)
            data.iloc[data_index[0], data.columns.get_loc("feature")] = gen_element[1]
            count += 1
            print(count)
        
        if count == 100:#50: # Prevents hours of running code and disconnection problems
            break
    return data

text_data = findVisualFeature(text_data, image_feature)
```

```python
# Remove the tuple type
def untuple(tup):
    return tup[0]

print(text_data.feature[0])
print([untuple(X) for X in text_data.feature[0]])
print(len([untuple(X) for X in text_data.feature[0]]))
print(len(text_data))
```

```python
# Initialisation
n_rows = 1000

def vis_data_to_df(data, n_rows):
    # Initialisation
    n_cols = 4096
    new_data = pd.DataFrame(np.zeros((n_rows, n_cols)))
    # Fill the new data
    for index, row in data.iterrows():
        #print(index, type(row.feature))
        if type(row.feature) != list:
            continue
        else: 
            new_data.iloc[index] = [untuple(X) for X in row.feature] #data.feature[index]]
    return new_data


train_vis_data = vis_data_to_df(text_data, n_rows)
print(train_vis_data.iloc[0, 0])
train_vis_data.head(10)
```

```python
# Save the results
train_vis_data.to_csv("vis_data_train.vis", header = None, index = None, sep = " ")
```
