# Training data

```python
cd task4_en-fr_training/
```

```python
from glob import iglob

for pathname in sorted(iglob("doc000*/source.segments")):
    print(pathname)
    f = open(pathname, "r")
    print(f.read() + "#doc#")
```

```python
# The training source
with open('../train.src', 'w') as outfile:
    for pathname in sorted(iglob("doc*/source.segments")):
        with open(pathname) as infile:
            outfile.write(infile.read() + "#doc#\n")
```

```python
# The training translation
with open('../train.mt', 'w') as outfile:
      for pathname in sorted(iglob("doc*/mt.segments")):
            with open(pathname) as infile:
                outfile.write(infile.read() + "#doc#\n")
```

```python
# The training score
with open('../train.mqm', 'w') as outfile:
      for pathname in sorted(iglob("doc*/document_mqm")):
            with open(pathname) as infile:
                  outfile.write(infile.read())
```
