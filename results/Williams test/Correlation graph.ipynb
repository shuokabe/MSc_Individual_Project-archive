```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Document-level biRNN

```python
# Load the table
document_corr = pd.read_csv('Document-level BiRNN/document-level-pred-pearson-sig.csv', sep='\t', index_col=0, na_values=" -")
print('Shape of the correlation table:', document_corr.shape)
document_corr.head()
```

```python
# Create a mask for the lower-triangle of the matrix
doc_mask_id = np.tril_indices(document_corr.shape[0])
doc_mask = np.zeros(document_corr.shape, dtype=np.bool)
doc_mask[doc_mask_id] = True
print(doc_mask)
```

```python
# Plot the correlation matrix
plt.figure(figsize=document_corr.shape)
sns.set_style('dark') # For a grey background in the heatmap

sns.heatmap(document_corr, vmin=0, cmap='coolwarm', mask=doc_mask, annot=True)
```
